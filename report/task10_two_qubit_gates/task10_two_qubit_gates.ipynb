```python
import json
from pathlib import Path
import sys

import matplotlib.pyplot as plt
import numpy as np

from musiq.analysis.common.state_utils import population_series
from musiq.workflow import create_model


MODEL_DIR = Path('.')
FIG_DIR = MODEL_DIR / "figures"
```

```python
model = create_model(
    circuits={
        "CZ": MODEL_DIR / "circuits" / "cz.yaml",
        "iSWAP": MODEL_DIR / "circuits" / "iswap.yaml",
    },
    solvers=MODEL_DIR / "solver.yaml",
    devices=MODEL_DIR / "device.yaml",
    pulses=MODEL_DIR / "pulse.yaml",
    analysers=MODEL_DIR / "analyser.yaml",
)
# model.run_all()

model.calibrate()
```

```python
model.run_all()
```

```python
from musiq.visualization import plot_case_metrics, plot_pulse, plot_sweep_metrics, plot_case_final_population

run_id = 'CZ'

fig = plt.Figure(figsize=(8,4))
gs = fig.add_gridspec(2, 1, height_ratios=[1.0, 2.5])
ax = fig.add_subplot(gs[0,0])
plot_pulse(
    ax, model, run_id=run_id, sample_rate=1e10, extend_to_trajectory_end=True,
    title = f"Pulse sequences: {run_id} Gate"
)
ax = fig.add_subplot(gs[1,0])
plot_case_metrics(
    ax, model, 'case_1', time_unit='ns',
    series_keys=["000", "100", "001", "010","020"],
    title = f"Population evoution: {run_id} Gate"
)

fig.tight_layout()
fig
```
